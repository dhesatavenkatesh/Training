```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

sentences = [
     "AI is changing the world.",
     "Artificial Intelligence is transforming industries.",
     "I love playing cricket."
]
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
print("====Embedding Dimensions====")

for i, emp in enumerate(embeddings , start=1):
  print(f"Sentence {i}: {emp.shape}")

sim1_2 = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
sim1_3 = cosine_similarity([embeddings[0]], [embeddings[2]])[0][0]
sim2_3 = cosine_similarity([embeddings[1]], [embeddings[2]])[0][0]

print("====Similarity Scores====")
print(f"Similarity between Sentence 1 and Sentence 2: {sim1_2:.4f}")
print(f"Similarity between Sentence 1 and Sentence 3: {sim1_3:.4f}")
print(f"Similarity between Sentence 2 and Sentence 3: {sim2_3:.4f}")

print("\n === Sample Embedding Vector (First 10 Values) ===")
print(embeddings[0][:10])
```

Task 3: Build Semantic Search Engine

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

documents = [
 "Python is used for AI development",
 "Machine learning helps computers learn patterns",
 "Transformers power modern LLMs",
 "Cloud computing provides scalable infrastructure",
 "Databases store application data"
]
model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = model.encode(documents)

print("Semantic Start Engine Ready!")
print("-"*40)

query = input("Enter your query:")
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, doc_embeddings)[0]
print("====Similarity Scores====")
for doc, scores in zip(documents, similarities):
  print(f"{doc}: {scores:.4f}")
```

```python
from transformers import pipeline
print("="*40)
print("Text Classification")
print("="*40)

classifier = pipeline(
    "sentiment-analysis",
    model = "distilbert-base-uncased-finetuned-sst-2-english"
)
text = "I love learning Artificial Intelligence."
result = classifier(text)
print(" Model Name:", " distilbert-base-uncased-finetuned-sst-2-english")
print(" Input     :", text)
print(" Output    :", result[0]['label'])
print(" Score     :", round(result[0]['score'],4))

print("\n"+"="*40)
print("Named Entity Recognition (NER)")
print("="*40)

ner = pipeline(
    "ner",
    aggregation_strategy = "simple",
    model = "dslim/bert-base-NER"
)
sentence = "Elon Musk founded SpaceX in California."
entities = ner(sentence)
print(" Model Name:", "dslim/bert-base-NER")
print("Input      :", sentence)
print("\n Detected Entities:")

for entity in entities:
  print(f"Entity   :{entity["word"]}")
  print(f"Type     :{entity["entity_group"]}")
  print(f"confidence   :{round(entity['score'],4)}")
  print('-'*30)

print("\n" + "="*40)
print("TEXT GENERATION")
print("="*40)

generator = pipeline(
    "text-generation",
    model="gpt2"
)
prompt = "The Future of AI is"
generated_text = generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']
print(" Model Name:", "gpt2")
print("Input      :", prompt)
print("Output     :",generated_text)

print("\nText Generation models do not provide a direct confidence score.")
```

Task 5: Tokenization Experiment

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
short_sentence = "Artificial Intelligence is amazing."
long_sentence = "Artificial Intelligence has revolutionized industries like healthcare, finance, and transportation."

programming_code = """
def add(a,b):
  return a + b

result = add(10,20)
print("result")
"""
texts = {
    "short_sentence": short_sentence,
    "long_sentence": long_sentence,
    "Programming code": programming_code
}

print("="*40)
print("Tokenization Experiment")
print("="*40)

tokens_counts ={ }
for name, text in texts.items():
  print(f"\n{name}")
  print('='*40)

  tokens = tokenizer.tokenize(text)
  tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
  decoded_text = tokenizer.decode(tokens_ids)

  tokens_counts[name] = len(tokens)

  print("Original Text:")
  print(text)

  print("\nTokens:")
  print(tokens)

  print("\nToken IDs:")
  print(tokens_ids)

  print("\nDecoded Text:")
  print(decoded_text)

  print("\nToken Count:", len(tokens))

print("\n" + "=" * 60)
print("TOKEN COUNT COMPARISON")
print("=" * 60)

for text_type, count in tokens_counts.items():
    print(f"{text_type}: {count} tokens")

```

Task 6: Build Mini Question Answering System

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased-distilled-squad")
model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad")

context = """
Python is a programming language created by Guido van Rossum.
It is widely used in AI, web development, automation, and data science.
"""

questions = [
    "Who created Python?",
    "Where is Python used?"
]

print("=" * 60)
print("MINI QUESTION ANSWERING SYSTEM")
print("=" * 60)

print("\nContext:")
print(context)

for question in questions:
    # Tokenize the input
    inputs = tokenizer(question, context, return_tensors="pt", truncation=True)

    # Get model outputs
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the most likely start and end indices
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits

    answer_start = torch.argmax(answer_start_scores)
    answer_end = torch.argmax(answer_end_scores) + 1 # +1 because slicing is exclusive

    # Convert tokens to answer string
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))

    # Calculate a simple confidence score (e.g., product of softmax scores)
    # This is a very simple way; actual pipeline might use more sophisticated methods
    start_prob = torch.softmax(answer_start_scores, dim=1)[0, answer_start].item()
    end_prob = torch.softmax(answer_end_scores, dim=1)[0, answer_end - 1].item()
    confidence_score = start_prob * end_prob

    print("\nQuestion :", question)
    print("Answer   :", answer)
    print("Confidence Score :", round(confidence_score, 4))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

model = SentenceTransformer('all-MiniLM-L6-v2')

documents = [
    "Python is a popular programming language.",
    "Machine learning is a subset of artificial intelligence.",
    "Deep learning uses neural networks.",
    "Data science combines statistics and computing.",
    "Natural Language Processing enables machines to understand text.",
    "Computer vision helps machines interpret images.",
    "Cloud computing provides scalable resources.",
    "Cybersecurity protects systems from attacks.",
    "Big data analytics extracts insights from data.",
    "SQL is used to manage relational databases.",
    "Java is an object-oriented programming language.",
    "JavaScript powers interactive web applications.",
    "HTML structures web pages.",
    "CSS styles websites.",
    "Git is a version control system.",
    "Docker enables containerization.",
    "Kubernetes manages containers.",
    "Linux is an open-source operating system.",
    "Artificial Intelligence is transforming industries.",
    "Reinforcement learning learns through rewards.",
    "Supervised learning uses labeled data.",
    "Unsupervised learning finds hidden patterns.",
    "Random Forest is an ensemble learning algorithm.",
    "XGBoost is a powerful boosting algorithm.",
    "Pandas is a Python library for data analysis.",
    "NumPy supports numerical computing.",
    "Matplotlib is used for data visualization.",
    "TensorFlow is a deep learning framework.",
    "PyTorch is widely used in AI research.",
    "Transformers are powerful NLP models.",
    "BERT is a transformer-based language model.",
    "GPT generates human-like text.",
    "Chatbots use NLP techniques.",
    "Recommendation systems suggest relevant content.",
    "Data preprocessing improves model performance.",
    "Feature engineering creates useful predictors.",
    "Cross-validation evaluates model stability.",
    "Regression predicts continuous values.",
    "Classification predicts categories.",
    "Clustering groups similar data points.",
    "Decision trees are easy to interpret.",
    "Neural networks mimic brain structures.",
    "Generative AI creates new content.",
    "Large Language Models understand language.",
    "Data visualization communicates insights.",
    "Business intelligence supports decision making.",
    "Automation reduces manual work.",
    "Web development creates websites.",
    "Mobile applications run on smartphones.",
    "Software testing improves quality."
]

document_embeddings = model.encode(documents)

def semantic_search(query, top_k=5):

    query_embedding = model.encode([query])

    similarities = cosine_similarity(
        query_embedding,
        document_embeddings
    )[0]

    top_indices = similarities.argsort()[::-1][:top_k]

    results = []

    for idx in top_indices:
        results.append({
            "Document": documents[idx],
            "Similarity Score": round(float(similarities[idx]), 4)
        })

    return pd.DataFrame(results)

query = input("Enter your search query: ")

results = semantic_search(query)

print("\nTop Matching Documents")
print("=" * 60)
print(results.to_string(index=False))
```
