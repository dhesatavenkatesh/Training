Task 6: Agent Communication Framework

```python
from datetime import datetime
from collections import defaultdict

class Message:
  """
  Represent a message exchanged between agents.
  """
  def __init__(self, sender, recevier, message_type, content):
    self.sender = sender
    self.recevier = recevier
    self.message_type = message_type
    self.content = content
    self.timestamp = datetime.now()

  def __str__(self):
    return (
        f"[{self.timestamp.strftime('%H:%M:%S')}]"
        f"{self.sender}->{self.recevier}"
        f"({self.message_type}): {self.content}"
    )

class CommunicationLogger:
  """
  Stores  all communication logs.
  """
  def __init__(self):
    self.logs = []
  def log(self, message):
    self.logs.append(str(message))
    print(str(message))

  def get_logs(self):
    return self.logs

class MessageBus:
    """
    Central communication hub for agents.
    """

    def __init__(self):
        self.agents = {}
        self.subscribers = defaultdict(list)
        self.logger = CommunicationLogger()

    def register_agent(self, agent):
        self.agents[agent.name] = agent
        agent.message_bus = self

    def send_message(
        self,
        sender,
        receiver,
        message_type,
        content
    ):
        message = Message(
            sender,
            receiver,
            message_type,
            content
        )

        self.logger.log(message)

        if receiver in self.agents:
            self.agents[receiver].receive_message(message)

    def assign_task(self, sender, receiver, task):
        self.send_message(
            sender,
            receiver,
            "TASK_ASSIGNMENT",
            task
        )

    def route_response(
        self,
        sender,
        receiver,
        response
    ):
        self.send_message(
            sender,
            receiver,
            "TASK_RESPONSE",
            response
        )

    def subscribe(self, event_name, agent_name):
        self.subscribers[event_name].append(agent_name)

    def publish_event(
        self,
        sender,
        event_name,
        data
    ):
        subscribers = self.subscribers.get(
            event_name,
            []
        )

        for agent in subscribers:
            self.send_message(
                sender,
                agent,
                f"EVENT:{event_name}",
                data
            )


class Agent:
    """
    Base Agent
    """

    def __init__(self, name):
        self.name = name
        self.message_bus = None

    def receive_message(self, message):
        print(
            f"{self.name} received: "
            f"{message.content}"
        )

    def send_message(
        self,
        receiver,
        message_type,
        content
    ):
        self.message_bus.send_message(
            self.name,
            receiver,
            message_type,
            content
        )

class ManagerAgent(Agent):

    def start_project(self, task):
        self.message_bus.assign_task(
            self.name,
            "ResearchAgent",
            task
        )

    def receive_message(self, message):
        print(
            f"\nManager received final response:\n"
            f"{message.content}"
        )


class ResearchAgent(Agent):

    def receive_message(self, message):

        if message.message_type == "TASK_ASSIGNMENT":

            research_result = (
                f"Research completed for: "
                f"{message.content}"
            )

            self.message_bus.route_response(
                self.name,
                "ArchitectAgent",
                research_result
            )


class ArchitectAgent(Agent):

    def receive_message(self, message):

        if message.message_type == "TASK_RESPONSE":

            architecture = (
                f"Architecture designed from "
                f"research -> {message.content}"
            )

            self.message_bus.route_response(
                self.name,
                "DeveloperAgent",
                architecture
            )


class DeveloperAgent(Agent):

    def receive_message(self, message):

        if message.message_type == "TASK_RESPONSE":

            implementation = (
                f"Implementation completed using "
                f"{message.content}"
            )

            self.message_bus.route_response(
                self.name,
                "ManagerAgent",
                implementation
            )

            self.message_bus.publish_event(
                self.name,
                "PROJECT_COMPLETED",
                implementation
            )

if __name__ == "__main__":

    bus = MessageBus()

    manager = ManagerAgent("ManagerAgent")
    research = ResearchAgent("ResearchAgent")
    architect = ArchitectAgent("ArchitectAgent")
    developer = DeveloperAgent("DeveloperAgent")

    bus.register_agent(manager)
    bus.register_agent(research)
    bus.register_agent(architect)
    bus.register_agent(developer)

    # Subscribe agents to events
    bus.subscribe(
        "PROJECT_COMPLETED",
        "ManagerAgent"
    )

    print("\n=== Starting Workflow ===\n")

    manager.start_project(
        "Build AI-powered travel planner"
    )

    print("\n=== Communication Logs ===\n")

    for log in bus.logger.get_logs():
        print(log)
```

Task 8: Multi-Agent Evaluation Framework

```python
import time
import statistics
from datetime import datetime
class MultiAgentEvaluator:
    """
    Evaluates Multi-Agent System Performance
    """
    def __init__(self):
        self.metrics = {
            "agent_accuracy": [],
            "communication_efficiency": [],
            "task_completion": [],
            "hallucination_rate": [],
            "collaboration_score": [],
            "latency": []
        }
    def record_accuracy(self, score):
        self.metrics["agent_accuracy"].append(score)
    def record_communication_efficiency(self, score):
        self.metrics["communication_efficiency"].append(score)
    def record_task_completion(self, score):
        self.metrics["task_completion"].append(score)
    def record_hallucination_rate(self, score):
        self.metrics["hallucination_rate"].append(score)
    def record_collaboration_score(self, score):
        self.metrics["collaboration_score"].append(score)
    def record_latency(self, latency):
        self.metrics["latency"].append(latency)
    def calculate_average(self, metric_name):
        values = self.metrics[metric_name]
        if not values:
            return 0
        return round(statistics.mean(values), 2)
    def overall_score(self):
        accuracy = self.calculate_average(
            "agent_accuracy"
        )
        communication = self.calculate_average(
            "communication_efficiency"
        )
        completion = self.calculate_average(
            "task_completion"
        )
        hallucination = self.calculate_average(
            "hallucination_rate"
        )
        collaboration = self.calculate_average(
            "collaboration_score"
        )
        final_score = (
            accuracy * 0.30 +
            communication * 0.15 +
            completion * 0.25 +
            collaboration * 0.20 +
            (100 - hallucination) * 0.10
        )
        return round(final_score, 2)

    def generate_report(
        self,
        filename="evaluation_report.md"
    ):
        report = f"""
# Multi-Agent Evaluation Report
Generated: {datetime.now()}
---
## Performance Metrics
| Metric | Score |
|----------|----------|
| Agent Accuracy | {self.calculate_average("agent_accuracy")} % |
| Communication Efficiency | {self.calculate_average("communication_efficiency")} % |
| Task Completion | {self.calculate_average("task_completion")} % |
| Hallucination Rate | {self.calculate_average("hallucination_rate")} % |
| Collaboration Score | {self.calculate_average("collaboration_score")} % |
| Average Latency | {self.calculate_average("latency")} sec |

---
## Analysis
### Agent Accuracy
Measures correctness of outputs.

### Communication Efficiency
Measures quality and effectiveness of message exchange.

### Task Completion
Measures successful completion of assigned tasks.

### Hallucination Rate
Measures frequency of incorrect or fabricated information.

### Collaboration Score
Measures cooperation among agents.

### Latency
Average response time.
---
## Overall System Score
**{self.overall_score()} / 100**
---
## Recommendations
"""
        if self.calculate_average("hallucination_rate") > 15:
            report += (
                "- Reduce hallucinations using validation layers\n"
            )
        if self.calculate_average("latency") > 5:
            report += (
                "- Optimize agent response times\n"
            )
        if self.calculate_average(
                "communication_efficiency") < 80:
            report += (
                "- Improve communication protocol\n"
            )
        if self.overall_score() > 90:
            report += (
                "- System performing exceptionally well\n"
            )
        with open(filename, "w", encoding="utf-8") as f:
            f.write(report)
        print(
            f"\nEvaluation report generated: {filename}"
        )
    def print_summary(self):
        print("\n========== EVALUATION SUMMARY ==========")
        print(
            f"Accuracy: "
            f"{self.calculate_average('agent_accuracy')}%"
        )
        print(
            f"Communication: "
            f"{self.calculate_average('communication_efficiency')}%"
        )
        print(
            f"Task Completion: "
            f"{self.calculate_average('task_completion')}%"
        )
        print(
            f"Hallucination Rate: "
            f"{self.calculate_average('hallucination_rate')}%"
        )
        print(
            f"Collaboration: "
            f"{self.calculate_average('collaboration_score')}%"
        )
        print(
            f"Latency: "
            f"{self.calculate_average('latency')} sec"
        )
        print(
            f"Overall Score: "
            f"{self.overall_score()}/100"
        )

if __name__ == "__main__":
    evaluator = MultiAgentEvaluator()

    evaluator.record_accuracy(94)
    evaluator.record_accuracy(92)
    evaluator.record_accuracy(96)

    evaluator.record_communication_efficiency(88)
    evaluator.record_communication_efficiency(91)

    evaluator.record_task_completion(100)
    evaluator.record_task_completion(95)

    evaluator.record_hallucination_rate(5)
    evaluator.record_hallucination_rate(4)

    evaluator.record_collaboration_score(93)
    evaluator.record_collaboration_score(95)

    evaluator.record_latency(1.8)
    evaluator.record_latency(2.1)

    evaluator.print_summary()

    evaluator.generate_report()
```

```python
import os
from crewai import Agent, Task, Crew, Process, LLM

os.environ["GROQ_API_KEY"] = "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"

llm = LLM(
    model="groq/llama-3.3-70b-versatile",
    temperature=0.2
)
hr_agent = Agent(
    role="HR Agent",
    goal="Handle HR-related employee queries",
    backstory="""
    Expert in employee management, leave policies,
    recruitment, onboarding and HR compliance.
    """,
    llm=llm,
    verbose=True
)
payroll_agent = Agent(
    role="Payroll Agent",
    goal="Handle salary and payroll inquiries",
    backstory="""
    Payroll specialist responsible for
    compensation, taxation and payslips.
    """,
    llm=llm,
    verbose=True
)
customer_support_agent = Agent(
    role="Customer Support Agent",
    goal="Resolve customer issues effectively",
    backstory="""
    Experienced support specialist capable of
    troubleshooting and client communication.
    """,
    llm=llm,
    verbose=True
)
project_manager_agent = Agent(
    role="Project Management Agent",
    goal="Track project progress and delivery",
    backstory="""
    Agile project management specialist.
    Handles timelines, risks and status reporting.
    """,
    llm=llm,
    verbose=True
)
documentation_agent = Agent(
    role="Technical Documentation Agent",
    goal="Create and maintain documentation",
    backstory="""
    Technical writer specializing in
    architecture documents, SOPs and manuals.
    """,
    llm=llm,
    verbose=True
)
analytics_agent = Agent(
    role="Analytics Agent",
    goal="Generate business insights and reports",
    backstory="""
    Expert in KPIs, dashboards,
    performance analytics and reporting.
    """,
    llm=llm,
    verbose=True
)
validation_agent = Agent(
    role="Validation Agent",
    goal="Validate responses before delivery",
    backstory="""
    Senior QA reviewer ensuring
    completeness, accuracy and consistency.
    """,
    llm=llm,
    verbose=True
)
class AgentRouter:
    @staticmethod
    def select_agent(user_request):
        query = user_request.lower()
        if any(word in query for word in [
            "leave",
            "employee",
            "recruitment",
            "hr",
            "attendance"
        ]):
            return hr_agent
        elif any(word in query for word in [
            "salary",
            "payroll",
            "payslip",
            "tax",
            "bonus"
        ]):
            return payroll_agent
        elif any(word in query for word in [
            "customer",
            "support",
            "issue",
            "ticket",
            "client"
        ]):
            return customer_support_agent
        elif any(word in query for word in [
            "project",
            "status",
            "sprint",
            "roadmap",
            "delivery"
        ]):
            return project_manager_agent
        elif any(word in query for word in [
            "documentation",
            "manual",
            "api",
            "technical"
        ]):
            return documentation_agent
        elif any(word in query for word in [
            "analytics",
            "report",
            "dashboard",
            "metrics",
            "kpi"
        ]):
            return analytics_agent
        return customer_support_agent

class BlackRothAICrew:
    def __init__(self):
        self.validation_agent = validation_agent
    def process_request(self, user_request):
        selected_agent = AgentRouter.select_agent(
            user_request
        )
        print(
            f"\nSelected Agent: "
            f"{selected_agent.role}"
        )
        specialist_task = Task(
            description=f"""
            User Request:
            {user_request}
            Provide a complete response.
            """,
            expected_output="Detailed Response",
            agent=selected_agent
        )
        validation_task = Task(
            description="""
            Review the generated response.

            Validate:
            - Accuracy
            - Completeness
            - Clarity
            - Consistency

            Return final response.
            """,
            expected_output="Validated Final Response",
            agent=self.validation_agent,
            context=[specialist_task]
        )

        crew = Crew(
            agents=[
                selected_agent,
                self.validation_agent
            ],
            tasks=[
                specialist_task,
                validation_task
            ],
            process=Process.sequential,
            verbose=True
        )

        result = crew.kickoff()

        return result
if __name__ == "__main__":
    crew = BlackRothAICrew()
    requests = [
        "Show employee leave policy",
        "Generate payroll report for March",
        "Customer cannot login to portal",
        "Provide project sprint status",
        "Create API documentation",
        "Generate KPI dashboard report"
    ]
    for req in requests:
        print("\n" + "=" * 80)
        print("USER REQUEST")
        print("=" * 80)
        print(req)
        result = crew.process_request(req)
        print("\nFINAL RESPONSE")
        print("-" * 80)
        print(result)
```
