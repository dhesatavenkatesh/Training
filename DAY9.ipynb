Task 5: Build Sequential AI Workflow

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("GROQ_API_KEY")
client = Groq(api_key=api_key)

idea = input("Enter the Product idea:")

def ask_groq(prompt):
  response = client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      messages=[
          {
              "role":"user",
              "content":prompt
          }
      ]
  )
  return response.choices[0].message.content
vision = ask_groq(
    f"Generate Product Vision for Idea: {idea}"
)
market = ask_groq(
    f"Analyze market opportunities:\n{vision}"
)
stories = ask_groq(
    f"Generate User Stories:\n{vision}"
)
architecture = ask_groq(
    f"Generate Technical Architecture:\n{vision}"
)
roadmap = ask_groq(
    f"Generate development roadmap:\n{vision}"
)
print("\nProduct Vision:\n")
print(vision)
print("\nMarket Analysis:\n")
print(market)
print("\nUser Stories:\n")
print(stories)
print("\nTechnical Architecture:\n")
print(architecture)
print("\Roadmap:\n")
print(roadmap)
````

Task 6: Output Parsing & Structured

````python
import os
import json
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

feature = input("Enter feature name: ")

prompt = f"""
Create a user story for {feature}.

Return ONLY valid JSON.

Example:
{{
    "title": "User Login",
    "description": "User can login securely",
    "acceptance_criteria": [
        "Email validation",
        "Password verification",
        "Error handling"
    ]
}}
"""

try:
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    result = response.choices[0].message.content.strip()

    result = result.replace("```json", "")
    result = result.replace("```", "")
    result = result.strip()

    print("\nRaw Response:\n")
    print(result)

    parsed = json.loads(result)

    print("\nParsed JSON:\n")
    print(json.dumps(parsed, indent=4))

except json.JSONDecodeError as e:
    print("\nJSON Error:")
    print(e)

except Exception as e:
    print("\nError:")
    print(e)
````

Task 7: Conversation Memory

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("GROQ_API_KEY")
client = Groq(api_key=api_key)

history = []

print("Type exit to Quit")
print("Type reset to clear memory\n")

while True:
  user_input = input("You:")
  if user_input.lower() == "exit":
    break

  if user_input.lower() == "reset":
    history = []
    print("Memory Cleared")
    continue

  history.append(
      {
          "role":"user",
          "content":user_input
      }
  )
  response = client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      messages=history
  )

  answer = response.choices[0].message.content
  history.append(
      {
          "role":"assistant",
          "content":answer
      }
  )
  print("\n Ai",answer)
````

Task 8: Build a Domain AI Assistant

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("GROQ_API_KEY")
client = Groq(api_key=api_key)

history = [
    {
        "role":"system",
        "content":"""
        you are an expert software Development Assistant.
        Responsbilities:
        1.Code Review
        2.Architecture Design
        3.BUg Analysis
        4.Security Reviews
        5.Best Practices

        Always Provide structured responses.

        Response Format:
        Summary:
        Bugs:
        Performances Issues:
        Security Concerns:
        Recommdations:
        """
    }
]
templates={
    "1":"""
    Review the following code:
{input}

provide:
1.Bugs
2.Performance Issue
3.Security concerns
4.Best pratices
""",
    "2":"""
    Design Software Artiecture for :
    {input}
    Include :
    -Components
    -Database Design
    -API's
    -Deploment Strategy
    """,

    "3":"""
    Analyze the bug:
    {input}

    provide:
    - Root cause
    - fix
    - Prevention Strategy
    """

}
print("===Software development assistant===")
while True:

  print("\n 1. Code Review")
  print(" 2. Architecture Design")
  print(" 3. Bug Analysis")
  print(" 4. Exit")

  choice = input("Enter your choice:")
  if choice == "4":
    print("GoodBye!")
    break
  if choice not in templates:
    print("Invalid Option")
    continue
  user_input = input("\n Enter Details:\n")
  try:
    prompt = templates[choice].format(
        input=user_input
    )
    history.append(
        {
            "role":"user",
            "content":prompt
        }
    )
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=history
    )
    answer = response.choices[0].message.content
    history.append(
        {
            "role":"assistant",
            "content":answer
        }
    )
    print("\n" + "=" * 50)
    print(answer)
    print("=" * 50)
  except Exception as e:
    print("Error",e)
````

Bonus Challenge - AI Software Architect Assistant

````python
import os
from dotenv import load_dotenv
from groq import Groq
from langchain_core.prompts import PromptTemplate

load_dotenv()

client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

def ask_groq(prompt):

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.5
    )

    return response.choices[0].message.content

requirements_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
You are a Senior Business Analyst.

Project:
{project}

Generate:

1. Business Requirements
2. Functional Requirements
3. Non Functional Requirements

Use clear bullet points.
"""
)

user_story_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Generate 10 user stories.

Format:

As a <user>,
I want <goal>,
So that <benefit>.
"""
)

database_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Design database schema.

Include:

1. Tables
2. Columns
3. Primary Keys
4. Foreign Keys
5. Relationships
"""
)

api_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Generate REST APIs.

Include:

- Endpoint
- Method
- Purpose
- Request
- Response
"""
)

architecture_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Design Microservices Architecture.

Include:

1. Services
2. Responsibilities
3. Communication
4. Security
5. Scalability
"""
)

tech_stack_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Recommend technology stack.

Include:

- Frontend
- Backend
- Database
- Cache
- Cloud
- DevOps
"""
)

timeline_prompt = PromptTemplate(
    input_variables=["project"],
    template="""
Project:
{project}

Create development roadmap.

Include:

Phase
Duration
Deliverables
"""
)

print("=" * 70)
print("AI SOFTWARE ARCHITECT ASSISTANT")
print("=" * 70)

project = input("\nEnter Project Name: ")

try:

    print("\nGenerating Requirements Analysis...")
    requirements = ask_groq(
        requirements_prompt.format(project=project)
    )

    print("Generating User Stories...")
    user_stories = ask_groq(
        user_story_prompt.format(project=project)
    )

    print("Generating Database Design...")
    database_design = ask_groq(
        database_prompt.format(project=project)
    )

    print("Generating API List...")
    api_list = ask_groq(
        api_prompt.format(project=project)
    )

    print("Generating Architecture...")
    architecture = ask_groq(
        architecture_prompt.format(project=project)
    )

    print("Generating Technology Stack...")
    tech_stack = ask_groq(
        tech_stack_prompt.format(project=project)
    )

    print("Generating Development Timeline...")
    timeline = ask_groq(
        timeline_prompt.format(project=project)
    )

    print("\n" + "=" * 90)
    print(f"PROJECT : {project}")
    print("=" * 90)

    print("\n1. REQUIREMENTS ANALYSIS")
    print("-" * 90)
    print(requirements)

    print("\n2. USER STORIES")
    print("-" * 90)
    print(user_stories)

    print("\n3. DATABASE DESIGN")
    print("-" * 90)
    print(database_design)

    print("\n4. API LIST")
    print("-" * 90)
    print(api_list)

    print("\n5. MICROSERVICES ARCHITECTURE")
    print("-" * 90)
    print(architecture)

    print("\n6. TECHNOLOGY STACK")
    print("-" * 90)
    print(tech_stack)

    print("\n7. DEVELOPMENT TIMELINE")
    print("-" * 90)
    print(timeline)

    print("\n" + "=" * 90)
    print("REPORT GENERATED SUCCESSFULLY")
    print("=" * 90)

except Exception as e:
    print("\nError:", e)
````
