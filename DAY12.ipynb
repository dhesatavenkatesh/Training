Task 2: Build First Tool Calling Agent

```python
from datetime import datetime

class CalucatorTool:
  @staticmethod
  def calucate(expression):
    try:
      return eval(expression)
    except Exception as e:
      return f"Calculation Error:{e}"
class CurrentDataTool:
  @staticmethod
  def get_date():
    return datetime.now().strftime("%Y-%m-%d")
class TextSummaryTool:
  @staticmethod
  def summarize(text):
    sentences = text.split(".")
    summary = ".".join(sentences[:3])
    return summary.strip()
class ToolCallingAgent:
  def process(self,query):
    if any(op in query for op in ["+","-","*","/"]):
      return CalucatorTool.calucate(query)
    elif "date" in query.lower():
      return CurrentDataTool.get_date()
    elif len(query.split()) > 30:
      return TextSummaryTool.summarize(query)

    return "No suitable tool found"

if __name__=="__main__":
  agent = ToolCallingAgent()

  print(agent.process("250*45"))
  print(agent.process("current date"))
```

Task 3: Create Function Calling Framework

```python
class WheatherTool:
  @staticmethod
  def get_wheather(city):
    return{
        "city" :city,
        "temperature":"30'c",
        "condition":"sunny"
    }
class CurrencyTool:
  @staticmethod
  def convert(amount,from_currency,to_currency):
    rate = 83.0
    return{
        "amount":amount,
        "from":from_currency,
        "to":to_currency,
        "converted_amount":amount*rate
    }
class EmployeeLookupTool:
    employees = {
        "EMP001": {
            "id": "EMP001",
            "name": "Ajay",
            "department": "Engineering"
        },
        "EMP002": {
            "id": "EMP002",
            "name": "Priya",
            "department": "HR"
        }
    }

    @classmethod
    def lookup(cls, emp_id):
      return cls.employees.get(emp_id,"Employee not found")

def function_router(function_name, **kwargs):
    functions = {
        "weather":WheatherTool.get_wheather,
        "currency":CurrencyTool.convert,
        "employee_lookup":EmployeeLookupTool.lookup
    }
    if function_name in functions:
      return functions[function_name](**kwargs)

    return "Function not available"

if __name__ == "__main__":
  result =function_router(
      "employee_lookup",
      emp_id="EMP002"
  )
  print(result)
```

Task 4: Multi-Step Agent Workflow

```python
class MultiStepAgent:
  def generate_requirements(self):
    return [
        "User Regiration",
        "Driver Registration",
        "Ride Booking",
        "Payments",
        "Ride Tracking"
    ]

  def generate_user_stories(self):
    return [
        "As a rider I can book a rides",
        "As a Driver I can accepts rides",
        " As a user I can make payments"
    ]

  def generate_database_schema(self):
    return {
        "Users":["id","name","email"],
        "Drivers":["id","name","vehicle_type"],
        "Rides":["id","rider_id","driver_id"],
    }

  def generate_api_list(self):
    return [
            "POST /register",
            "POST /book-ride",
            "GET /ride-status",
            "POST /payment"
        ]

  def generate_roadmap(self):
    return [
            "Phase 1 - Requirements",
            "Phase 2 - Design",
            "Phase 3 - Development",
            "Phase 4 - Testing",
            "Phase 5 - Deployment"
        ]

  def execute(self):
     return {
            "Requirements": self.generate_requirements(),
            "User Stories": self.generate_user_stories(),
            "Database Schema": self.generate_database_schema(),
            "APIs": self.generate_api_list(),
            "Roadmap": self.generate_roadmap()
        }


if __name__ == "__main__":
    agent = MultiStepAgent()

    result = agent.execute()

    print("Requirements:")
    for item in result["Requirements"]:
        print("-", item)

    print("\nUser Stories:")
    for item in result["User Stories"]:
        print("-", item)

    print("\nDatabase Schema:")
    for table, columns in result["Database Schema"].items():
        print(f"{table}: {columns}")

    print("\nAPIs:")
    for api in result["APIs"]:
        print("-", api)

    print("\nRoadmap:")
    for phase in result["Roadmap"]:
        print("-", phase)
```

Task 6: Agent Memory System

```python
class AgentMemory:

    def __init__(self):
        self.long_term_memory = {}

    def remember_name(self, name):
        self.long_term_memory["name"] = name

    def get_name(self):
        return self.long_term_memory.get("name", "Unknown")


if __name__ == "__main__":

    memory = AgentMemory()

    user_input = input("User: ")

    if "my name is" in user_input.lower():
        name = user_input.split()[-1]
        memory.remember_name(name)

    question = input("User: ")

    if "what is my name" in question.lower():
        print(f"Agent: Your name is {memory.get_name()}.")
```

Task 7: Reflection & Self-Correction

```python
class ReflectionEngine:
    def initial_response(self, query):
        return (f"Initial answer for: {query}")

    def validate(self, response):
        issues = []
        if len(response) < 20:
            issues.append("Too short")
        if "unknown" in response.lower():
            issues.append("Missing information")
        return issues

    def self_correct(self, response, issues):
        corrected = response
        if issues:
            corrected += (  "\n\nImproved with additional details.")
        return corrected

    def process(self, query):
        response = self.initial_response(query)
        issues = self.validate(response)
        improved = self.self_correct(
            response,
            issues
        )
        return {
            "initial_response": response,
            "issues": issues,
            "improved_response": improved
        }

if __name__ == "__main__":
    engine = ReflectionEngine()
    result = engine.process("Explain AI Agents")
    print("Initial Response:")
    print(result["initial_response"])
    print("\nIssues Found:")
    if result["issues"]:
        for issue in result["issues"]:
            print("-", issue)
    else:
        print("No issues found")
    print("\nImproved Response:")
    print(result["improved_response"])
```

Task 8: Build a Multi-Agent System

```python
class ResearchAgent:
    def execute(self, request):
        return (f"Research Findings for {request}")

class BusinessAnalystAgent:
    def execute(self, research):
        return {
            "requirements": [
                "User Management",
                "Payments",
                "Notifications"
            ]
        }
class ArchitectAgent:
    def execute(self, requirements):
        return {
            "architecture":
                "Microservices + API Gateway"
        }
class ProjectManagerAgent:
    def execute(self, architecture):
        return {
            "roadmap": [
                "Planning",
                "Development",
                "Testing",
                "Release"
            ]
        }
class MultiAgentSystem:
    def run(self, request):
        research = ResearchAgent().execute(request)
        requirements = \
            BusinessAnalystAgent().execute(research)
        architecture = \
            ArchitectAgent().execute(requirements)
        roadmap = \
            ProjectManagerAgent().execute(architecture)
        return {
            "research": research,
            "requirements": requirements,
            "architecture": architecture,
            "roadmap": roadmap
        }
if __name__ == "__main__":
    system = MultiAgentSystem()
    result = system.run("Build Ride Booking App")
    print("Research Agent Output:")
    print(result["research"])
    print("\nBusiness Analyst Agent Output:")
    for req in result["requirements"]["requirements"]:
        print("-", req)
    print("\nArchitect Agent Output:")
    print(result["architecture"]["architecture"])
    print("\nProject Manager Agent Output:")
    for phase in result["roadmap"]["roadmap"]:
        print("-", phase)
```

Task 9: Agent Evaluation Framework

```python
import time
class AgentEvaluator:
    def evaluate(self):
        report = {
            "Task Completion Rate": "95%",
            "Accuracy": "92%",
            "Tool Usage": "90%",
            "Response Time": "1.2 sec",
            "Hallucination Rate": "3%",
            "User Satisfaction": "4.7/5"
        }
        with open(
            "agent_evaluation_report.md",
            "w"
        ) as file:
            file.write(
                "# Agent Evaluation Report\n\n"
            )
            for k, v in report.items():
                file.write(
                    f"- **{k}**: {v}\n"
                )
        return report
if __name__ == "__main__":
    evaluator = AgentEvaluator()
    result = evaluator.evaluate()
    print("Agent Evaluation Report")
    print("-" * 30)
    for metric, value in result.items():
        print(f"{metric}: {value}")
```
